Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x=[1,2,3,4,5,6,7]#weekdays
y=[50,51,52,48,47,49,46]#Temp info
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.title("Weather")
plt.xlabel("day")
plt.ylabel("temperature")
plt.plot(x,y,color='green',linewidth=2,linestyle='dashdot')
```

```python
plt.title("Weather")
plt.xlabel("day")
plt.ylabel("temperature")
plt.plot(x,y,color='green',linewidth=2,linestyle='dotted')
```

Axes Labels,Grid,Legend

```python
plt.xlabel("Days")
plt.ylabel("max_temp")
plt.plot(days,max_t)
```

Legend Label

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Day VS Temperature")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Average")
plt.legend(shadow=True)#shadow
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Day VS Temperature")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Average")
plt.legend(shadow=True,fontsize="large")#font of legend
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Day VS Temperature")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Average")
plt.legend(shadow=True)
plt.grid()
```
