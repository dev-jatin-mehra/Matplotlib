Pie Chart Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_val = [1400,600,300,410,250]
exp_place = ["Home Rent","Food","Bill Telecom","Car","Other Utilities"]
```

```python
plt.axis("equal")#if an ellipse is formed
plt.pie(exp_val,labels=exp_place,radius=1.5,autopct='%0.2F%%',shadow='enable')
plt.show()
```

```python
plt.axis("equal")#if an ellipse is formed
plt.pie(exp_val,labels=exp_place,radius=1.5,autopct='%0.2F%%',shadow='enable',explode=[0.1,0,0.2,0,0],startangle=180)
plt.show()
```
