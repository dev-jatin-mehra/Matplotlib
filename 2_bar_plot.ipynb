Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGL','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
plt.bar(company,revenue)
```

```python
xpos =  np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.xlabel("Company")
plt.ylabel("Revenue(Bn)")
plt.bar(xpos,revenue,label="Revenue")
plt.legend()
```

Creating A bar chart for profit and revenue

```python
plt.xticks(xpos,company)
plt.xlabel("Company")
plt.ylabel("Revenue(Bn)")
plt.bar(xpos-0.2,revenue,label="Revenue")
plt.bar(xpos+0.2,profit,label="Profit")
plt.legend()
```

Good Looking

```python
plt.xticks(xpos,company)
plt.xlabel("Company")
plt.ylabel("Revenue(Bn)")
plt.bar(xpos-0.2,revenue,label="Revenue",width=0.4)
plt.bar(xpos+0.2,profit,label="Profit", width=0.4)
plt.legend()
```

Horizontal Chart

```python
plt.yticks(xpos,company)
plt.title("US revenue stocks")
plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.title("US revenue stocks")
plt.barh(xpos-0.2,revenue,label="Revenue",height=0.4)
plt.barh(xpos+0.2,profit,label="Profit",height=0.4)
plt.legend()
```

Histograms

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,123,133,153,84,69,89,79,120,112,100]
```

```python
plt.hist(blood_sugar_men,bins=[80,100,125,150],rwidth=0.9)
```

```python
plt.hist(blood_sugar_men,bins=[80,100,125,150],rwidth=0.9,color='g')
```

```python
plt.hist(blood_sugar_men,bins=[80,100,125,150],rwidth=0.9,color='g',histtype='step')
```

```python
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.9,color=['blue','orange'])
```

```python
plt.xlabel("Sugar range")
plt.ylabel("Frequency")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=0.9,color=['blue','orange'],label=['men','women'])
plt.legend()
plt.savefig("histogram.png",pad_inches=2,bbox_inches="tight")#savefig function used to save chart to a file as a file
```
